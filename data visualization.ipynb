```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data_1 = pd.read_csv('https://assets.datacamp.com/production/repositories/558/datasets/5f4f1a9bab95fba4d7fea1ad3c30dcab8f5b9c96/percent-bachelors-degrees-women-usa.csv')
```

```python
data_1.head()
```

```python
#Analysing the percentage of physical science and computer science degrees awarded to women from 1970 to 2011 inclusive

year = data_1[['Year']]
physical_sciences = data_1['Physical Sciences'].values
computer_science = data_1['Computer Science'].values

plt.style.use('ggplot')
# A 2x2 subplot with left subplot active
plt.subplot(2,2,1)
plt.plot(year, physical_sciences, color='blue')
plt.title('Physical Sciences')

# right subplot active with 2x2 subplot grid
plt.subplot(2,2,2)
plt.plot(year, computer_science, color='red')
plt.title('Computer Science')

#improve the spacing between subplots 
plt.tight_layout()
plt.show()
```

```python
data_2 = pd.read_csv('https://assets.datacamp.com/production/repositories/558/datasets/1a03987ad77b38d61fc4c692bf64454ddf345fbe/auto-mpg.csv')
```

```python
data_2.head()
```

```python
#Using seaborn to visualize linear regression, residuals of a regression and higher order regression

sns.lmplot(x='weight', y='hp', data=data_2)
plt.title('Linear regression')
plt.xlabel('weight')
plt.ylabel('hp')
plt.show()
```

```python
# higher order regression
plt.scatter(data_2['weight'], data_2['mpg'], label='data', color='red', marker='o')

#linear regression of order 1 which is linear between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=data_2, scatter=None, color='blue', label='order 1')

# linear regression of order 2 which is of the form of a polynomial between 'weight' and 'mpg'
#This captures more points unlike the linear
sns.regplot(x='weight', y='mpg', data=data_2, scatter=None, order=2, color='green', label='order 2')

# Add a legend and display the plot
plt.title('Higher order regression')
plt.legend(loc='upper right')
plt.show()
```
